```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets 
```

```python
x,y= datasets.make_blobs(n_samples=1500,n_features=2,
   random_state=10)
```

```python
y.shape
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
link =linkage(x,'average')
```

```python
label=range(1,16)
```

```python
dendrogram(link,orientation='top',labels=label,  distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
cluster=AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='average')
```

```python
cluster.fit_predict(x)
```

```python
labelss=cluster.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=labelss)
```
